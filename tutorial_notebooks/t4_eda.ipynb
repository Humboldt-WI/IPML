## Plot a histogram

```python
import matplotlib.pyplot as plt
```

```python
var = 'Battery capacity (%)'

plt.hist(data[var], bins=20)
plt.title(f'Histogram of {var}')
plt.xlabel(var)
plt.ylabel('Count')
plt.show()
```

## Python enumerate

```python
vars = [
    'Weight (grams)',
    'Actual Lease Duration (months)',
    'Battery capacity (%)',
]
```

```python
# simple iteration over a list
for i in vars:
    print(i)
```

```python
# enumerate gives us a two elements, the iteration number and the original element
for i in enumerate(vars):
    print(i)
```

```python
# we can apply return unpacking to split them up
for i, j in enumerate(vars):
    print(i, j)
```

## Matplotlib subplots

Putting it all together: Plotting three histograms side by side

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

for i, var in enumerate(vars):
    axes[i].hist(data[var], bins=20)
    axes[i].set_title(f'Histogram of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Count')
    axes[i].set_ylim(0, 1000)
    
plt.show()
```
